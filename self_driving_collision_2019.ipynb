```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#data ca autonomous-vehicle collision report 2019

company = ["GMCruise", "Waymo", "Lyft", "Zoox Inc.", 
           "Pony.ai", "Aurora In"]

ypos = np.arange(len(company))

collision = [41, 16, 6, 4, 3, 2]
```

```python
#plot 

plt.barh(ypos, collision, align='center', alpha=0.5, color='#9DB9AE')
plt.yticks(ypos, company)
plt.ylabel("Company")
plt.xlabel('# of self-driving collision')
plt.title("Reported self-driving collision in 2019")

#show graph
plt.savefig("bar.jpg", dpi=400)
plt.show()
```
